### Store the datasets in database tables

To analyze the data using SQL, it first needs to be loaded into SQLite DB.
We will create three tables in as under:

1.  **CENSUS_DATA**
2.  **CHICAGO_PUBLIC_SCHOOLS**
3.  **CHICAGO_CRIME_DATA**


Load the `pandas` and `sqlite3` libraries and establish a connection to `FinalDB.db`


```python
import pandas as pd
census = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DB0201EN-SkillsNetwork/labs/FinalModule_Coursera_V5/data/ChicagoCensusData.csv?utm_medium=Exinfluencer&utm_source=Exinfluencer&utm_content=000026UJ&utm_term=10006555&utm_id=NA-SkillsNetwork-Channel-SkillsNetworkCoursesIBMDeveloperSkillsNetworkDB0201ENSkillsNetwork20127838-2021-01-01')
schools = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DB0201EN-SkillsNetwork/labs/FinalModule_Coursera_V5/data/ChicagoPublicSchools.csv?utm_medium=Exinfluencer&utm_source=Exinfluencer&utm_content=000026UJ&utm_term=10006555&utm_id=NA-SkillsNetwork-Channel-SkillsNetworkCoursesIBMDeveloperSkillsNetworkDB0201ENSkillsNetwork20127838-2021-01-01')
crime = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DB0201EN-SkillsNetwork/labs/FinalModule_Coursera_V5/data/ChicagoCrimeData.csv?utm_medium=Exinfluencer&utm_source=Exinfluencer&utm_content=000026UJ&utm_term=10006555&utm_id=NA-SkillsNetwork-Channel-SkillsNetworkCoursesIBMDeveloperSkillsNetworkDB0201ENSkillsNetwork20127838-2021-01-01')
```

```python
import sqlite3
conn = sqlite3.connect('FinalDB.db')
```

Load the SQL magic module


```python
%load_ext sql
```

Use `Pandas` to load the data available in the links above to dataframes. Use these dataframes to load data on to the database `FinalDB.db` as required tables.


```python
census.to_sql('census_db', conn, if_exists='replace', index=False)
schools.to_sql('schools_db', conn, if_exists='replace', index=False)
crime.to_sql('crime_db', conn, if_exists='replace', index=False)
```

## Problems

Now write and execute SQL queries to solve assignment problems

### Problem 1

##### Find the total number of crimes recorded in the CRIME table.


```python
crime.count().max()
```

### Problem 2

##### List community area names and numbers with per capita income less than 11000.


```python
%sql select COMMUNITY_AREA_NUMBER, COMMUNITY_AREA_NAME from crime_db where PER_CAPITA_INCOME<11000

```

### Problem 3

##### List all case numbers for crimes involving minors?(children are not considered minors for the purposes of crime analysis) 


```python
# %sql select CASE_NUMBER from crime_db limit 1
%sql select CASE_NUMBER from (select case_number, LOWER(DESCRIPTION) as d from crime_db) where d like '%minor%'
```

### Problem 4

##### List all kidnapping crimes involving a child?


```python
%sql select CASE_NUMBER, d from (select case_number, LOWER(DESCRIPTION) as d from crime_db) where d like '%child%'
```

### Problem 5

##### List the kind of crimes that were recorded at schools. (No repetitions)


```python
cur = conn.cursor()
cur.execute('select * from crime_db where location_description like "%school%"')
pd.DataFrame(cur.fetchall())
```

### Problem 6

##### List the type of schools along with the average safety score for each type.


```python
schools.groupby('Elementary, Middle, or High School').count()
```

### Problem 7

##### List 5 community areas with highest % of households below poverty line


```python
%sql select COMMUNITY_AREA_NAME, COMMUNITY_AREA_NUMBER from census_db order by PERCENT_HOUSEHOLDS_BELOW_POVERTY desc limit 5
```

### Problem 8

##### Which community area is most crime prone? Display the coumminty area number only.


```python
%%sql
select COMMUNITY_AREA_NUMBER ,COUNT(COMMUNITY_AREA_NUMBER) as FREQUENCY
from crime_db 
group by COMMUNITY_AREA_NUMBER
order by COUNT(COMMUNITY_AREA_NUMBER) desc
limit 1
```

### Problem 9

##### Use a sub-query to find the name of the community area with highest hardship index


```python
cur.execute('''select community_area_name from census_db where hardship_index = (select max(hardship_index) from census_db)''')
cur.fetchall()
```

### Problem 10

##### Use a sub-query to determine the Community Area Name with most number of crimes?


```python
%%sql
select community_area_name from census_db 
where COMMUNITY_AREA_NUMBER = (select COMMUNITY_AREA_NUMBER from crime_db 
    group by COMMUNITY_AREA_NUMBER
    order by COUNT(COMMUNITY_AREA_NUMBER) DESC
    limit 1)
limit 1;
```
